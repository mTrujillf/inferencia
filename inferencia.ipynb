```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('mohoma.csv',encoding='latin-1')
```

```python
df['ap_1_3'] = df['ap_1_3'].replace({' ': '10'})
df['ap_1_3'] = df['ap_1_3'].replace({np.nan: '10'})
df['ap_1_3'] = df['ap_1_3'].astype(int)
df['ap_1'] = df['ap_1'].replace({' ': '2'})
df['ap_1'] = df['ap_1'].astype(int)
df_nuevo=df.loc[df['ap_1'] < 2]
df_nuevo=df_nuevo.loc[df_nuevo['ap_1_3'] <= 4]
```

```python
df_nuevo['ap_1_2'] = df_nuevo['ap_1_2'].replace({np.nan: '0'})
df_nuevo['ap_1_2'] = df_nuevo['ap_1_2'].replace({' ': '0'})
df_nuevo['ap_1_2'] = df_nuevo['ap_1_2'].astype(int)
df_nuevo = df_nuevo.loc[df_nuevo['ap_1_2'] > 0]
```

```python
def colTabla(df):
    res = []
    df_a = df['ap_1_3']
    for x in range(len(df_a)):
        val = df_a.iloc[x]
        if val == 1:
            res.append(12)
        else:
            if val == 2:
                res.append(6)
            else:
                if val == 3:
                    res.append(4)
                else:
                    res.append(1)
    return res

df_nuevo['periodo'] = colTabla(df_nuevo)
df_nuevo['pago_anual'] = df_nuevo['periodo'] * df_nuevo['ap_1_2']
df_nuevo
```

```python
df_nuevo['ap_6_1'] = df_nuevo['ap_6_1'].replace({np.nan: '0'})
df_nuevo['ap_6_1'] = df_nuevo['ap_6_1'].replace({' ': '0'})
df_nuevo['ap_6_1'] = df_nuevo['ap_6_1'].astype(int)
df_nuevo['ap_6_2'] = df_nuevo['ap_6_2'].replace({np.nan: '0'})
df_nuevo['ap_6_2'] = df_nuevo['ap_6_2'].replace({' ': '0'})
df_nuevo['ap_6_2'] = df_nuevo['ap_6_2'].astype(int)
df_nuevo['ap_6_3'] = df_nuevo['ap_6_3'].replace({np.nan: '0'})
df_nuevo['ap_6_3'] = df_nuevo['ap_6_3'].replace({' ': '0'})
df_nuevo['ap_6_3'] = df_nuevo['ap_6_3'].astype(int)
df_nuevo['ap_6_4'] = df_nuevo['ap_6_4'].replace({np.nan: '0'})
df_nuevo['ap_6_4'] = df_nuevo['ap_6_4'].replace({' ': '0'})
df_nuevo['ap_6_4'] = df_nuevo['ap_6_4'].astype(int)
df_nuevo['ap_6_5'] = df_nuevo['ap_6_5'].replace({np.nan: '0'})
df_nuevo['ap_6_5'] = df_nuevo['ap_6_5'].replace({' ': '0'})
df_nuevo['ap_6_5'] = df_nuevo['ap_6_5'].astype(int)
df_nuevo['ap_6_6'] = df_nuevo['ap_6_6'].replace({np.nan: '0'})
df_nuevo['ap_6_6'] = df_nuevo['ap_6_6'].replace({' ': '0'})
df_nuevo['ap_6_6'] = df_nuevo['ap_6_6'].astype(int)
df_nuevo['ap_6_7'] = df_nuevo['ap_6_7'].replace({np.nan: '0'})
df_nuevo['ap_6_7'] = df_nuevo['ap_6_7'].replace({' ': '0'})
df_nuevo['ap_6_7'] = df_nuevo['ap_6_7'].astype(int)
df_nuevo['ap_6_8'] = df_nuevo['ap_6_8'].replace({np.nan: '0'})
df_nuevo['ap_6_8'] = df_nuevo['ap_6_8'].replace({' ': '0'})
df_nuevo['ap_6_8'] = df_nuevo['ap_6_8'].astype(int)
df_nuevo['ap_6_9'] = df_nuevo['ap_6_9'].replace({np.nan: '0'})
df_nuevo['ap_6_9'] = df_nuevo['ap_6_9'].replace({' ': '0'})
df_nuevo['ap_6_9'] = df_nuevo['ap_6_9'].astype(int)
df_nuevo['ap_6_10'] = df_nuevo['ap_6_10'].replace({np.nan: '0'})
df_nuevo['ap_6_10'] = df_nuevo['ap_6_10'].replace({' ': '0'})
df_nuevo['ap_6_10'] = df_nuevo['ap_6_10'].astype(int)
df_nuevo['ap_6_11'] = df_nuevo['ap_6_11'].replace({np.nan: '0'})
df_nuevo['ap_6_11'] = df_nuevo['ap_6_11'].replace({' ': '0'})
df_nuevo['ap_6_11'] = df_nuevo['ap_6_11'].astype(int)
df_nuevo['ap_6_1'] = df_nuevo['ap_6_1'].replace({2: 0})
df_nuevo['ap_6_2'] = df_nuevo['ap_6_2'].replace({2: 0})
df_nuevo['ap_6_3'] = df_nuevo['ap_6_3'].replace({2: 0})
df_nuevo['ap_6_4'] = df_nuevo['ap_6_4'].replace({2: 0})
df_nuevo['ap_6_5'] = df_nuevo['ap_6_5'].replace({2: 0})
df_nuevo['ap_6_6'] = df_nuevo['ap_6_6'].replace({2: 0})
df_nuevo['ap_6_7'] = df_nuevo['ap_6_7'].replace({2: 0})
df_nuevo['ap_6_8'] = df_nuevo['ap_6_8'].replace({2: 0})
df_nuevo['ap_6_9'] = df_nuevo['ap_6_9'].replace({2: 0})
df_nuevo['ap_6_10'] = df_nuevo['ap_6_10'].replace({2: 0})
df_nuevo['ap_6_11'] = df_nuevo['ap_6_11'].replace({2: 0})
df_nuevo['tot'] = df_nuevo['ap_6_1'] + df_nuevo['ap_6_2'] + df_nuevo['ap_6_3'] + df_nuevo['ap_6_4'] + df_nuevo['ap_6_5'] + df_nuevo['ap_6_6'] + df_nuevo['ap_6_7'] + df_nuevo['ap_6_8'] + df_nuevo['ap_6_9'] + df_nuevo['ap_6_10'] + df_nuevo['ap_6_11']
```

```python
def calculaS2(df):
    n = len(df)
    m = np.mean(df)
    s = df - m
    s = np.dot(s,s)
    s = s/(n-1)
    
    return s

s2 = calculaS2(df_nuevo['pago_anual'])
s = pow(s2,1/2)
print(s)
```
